# Advanced Analytics + Risk Metrics

This notebook summarizes VaR/CVaR, rolling Sharpe, investor cohorts, SIP continuity, risk-appetite recommendations, and sector HHI concentration.

```python
from pathlib import Path
import pandas as pd

BASE_DIR = Path('..').resolve()
PROCESSED_DIR = BASE_DIR / 'data' / 'processed'
var_cvar = pd.read_csv(PROCESSED_DIR / 'var_cvar_report.csv')
cohort = pd.read_csv(PROCESSED_DIR / 'investor_cohort_analysis.csv')
sip_continuity = pd.read_csv(PROCESSED_DIR / 'sip_continuity_analysis.csv')
sector_hhi = pd.read_csv(PROCESSED_DIR / 'sector_hhi_concentration.csv')
var_cvar.head()
```

## Methodology

- Historical VaR 95%: 5th percentile of each fund's daily return distribution.
- CVaR 95%: average return for observations at or below the VaR threshold.
- Rolling Sharpe: `returns.rolling(90).mean() / returns.rolling(90).std() * sqrt(252)`.
- Investor cohorts: grouped by first transaction year.
- SIP continuity: investors with 6+ SIP transactions are at risk if any transaction gap exceeds 35 days.
- Sector HHI: sum of squared sector weights per equity fund.
