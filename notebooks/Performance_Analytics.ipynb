# Fund Performance Analytics

This notebook documents the Day 4 performance analytics workflow: daily returns, trailing CAGR, Sharpe, Sortino, alpha/beta, drawdown, scorecard ranking, and benchmark comparison.

```python
from pathlib import Path
import pandas as pd
import numpy as np
from scipy.stats import linregress

BASE_DIR = Path('..').resolve()
PROCESSED_DIR = BASE_DIR / 'data' / 'processed'
scorecard = pd.read_csv(PROCESSED_DIR / 'fund_scorecard.csv')
alpha_beta = pd.read_csv(PROCESSED_DIR / 'alpha_beta.csv')
scorecard.head()
```

## Methodology

- Daily return: `NAV_t / NAV_t-1 - 1` for each scheme.
- CAGR: `(NAV_end / NAV_start) ** (1 / observed_years) - 1` for 1-year, 3-year, and available-history 5-year windows.
- Sharpe: `(Rp - Rf) / Std(Rp) * sqrt(252)` with `Rf = 6.5%`.
- Sortino: same excess-return numerator, downside standard deviation denominator.
- Alpha/Beta: `scipy.stats.linregress` of fund returns on NIFTY100 returns; annualized alpha is `intercept * 252`.
- Scorecard: 30% 3-year return rank, 25% Sharpe rank, 20% alpha rank, 15% inverse expense-ratio rank, 10% inverse max-drawdown rank.
