## 10 Key EDA Findings


**Insight 1.** NAV compounding was broad-based, with ICICI Pru Midcap Fund - Regular - Growth ending the period about 248.7% above its 2022 base; see Chart 2.

**Insight 2.** The latest fund-house AUM snapshot is led by SBI Mutual Fund at Rs. 1,250,000 crore; see Chart 5.

**Insight 3.** SIP inflows reached their observed high of Rs. 31,002 crore in Dec 2025; see Chart 6.

**Insight 4.** Liquid attracted the highest cumulative net category inflow over the period; see Charts 8 and 9.

**Insight 5.** SIP ticket sizes vary meaningfully by age group, with visible upper-tail outliers across cohorts; see Chart 11.

**Insight 6.** Madhya Pradesh contributes the largest aggregate SIP transaction amount among states; see Chart 13.

**Insight 7.** City-tier participation is split T30: 66.3%, B30: 33.7%; see Chart 14.

**Insight 8.** Total folios grew by 97.0% across the available series; see Chart 15.

**Insight 9.** The selected top-AUM funds have an average pairwise daily-return correlation of 0.01; see Chart 16.

**Insight 10.** Banking is the largest aggregate equity-holding sector, while 14 schemes have expense ratios below 1%; see Charts 17 and 18.
